```python
df = pd.read_csv('day_wise.csv', index_col=False)
```

```python
df['date'] = pd.to_datetime(df['date'])
df = df.sort_values(by='date')
df
```

```python
df
```

```python
start = '07-01-2020'
end = '07-31-2020'
```

```python
mask = (df['date'] >= start) & (df['date'] <= end)
```

```python
df1 = df.loc[mask]
df1
```

```python
df1.sort_values(by='date')
```

```python
totp = int(df1['tot'].sum())
posp = int(df1['pos'].sum())
negp = int(df1['neg'].sum())
neutp = int(df1['neut'].sum())
avgp = df1['avg'].mean()


maxpos = df1.nlargest(1, 'p/n').iloc[0]['date'].date()
maxneg = df1.nsmallest(1, 'p/n').iloc[0]['date'].date()
```

```python
df_phase = pd.read_csv('phase_wise.csv')
```

```python
temp = {'phase': 'U2',
        'start': start,
        'end': end,
        'tot': totp,
       'avg': avgp,
       'pos': posp,
       'neg': negp,
       'neut': neutp,
        'date_mpos': maxpos,
        'date_mneg': maxneg
        }
df_phase = df_phase.append(temp, ignore_index=True)
df_phase
```

```python
df_phase.to_csv('phase_wise.csv', index=None)
```
